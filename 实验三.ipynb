```python
import numpy as np
import matplotlib.pyplot as plt

# 定义系统模型
def system_model(y, u, e, k):
    return 0.9 * y[k-1] + u[k-2] + e[k] - 0.7 * e[k-1]

# 仿真参数
n = 100  # 仿真长度
y = np.zeros(n)  # 输出信号
u = np.zeros(n)  # 输入信号
e = np.random.normal(0, 1, n)  # 白噪声

# 系统时延
m = 2

# 使用估计的参数进行最小方差调节
a1, a2, b0, b1 = 0.9, 0, 1, 0  # 假设系统模型已知
def min_variance_controller(y, u, k):
    u[k] = -(a1 * y[k-1] + a2 * y[k-2] - b1 * u[k-1]) / b0

# 计算控制输入和系统输出
for k in range(m, n):
    min_variance_controller(y, u, k)
    y[k] = system_model(y, u, e, k)

# 绘制系统输出
plt.plot(y, label='System Output y(k)')
plt.legend()
plt.xlabel('k')
plt.ylabel('y(k)')
plt.title('System Output with Min Variance Controller')
plt.show()
```

```python
# 计算控制误差
control_error = np.zeros(n)
for k in range(1, n):
    control_error[k] = y[k] - y[k-1]

# 绘制控制误差
plt.plot(control_error, label='Control Error e(k)')
plt.legend()
plt.xlabel('k')
plt.ylabel('e(k)')
plt.title('Control Error with Min Variance Controller')
plt.show()
```

```python
# 定义系统模型
def system_model(y, u, e, k):
    return 0.9 * y[k-1] + u[k-2] + e[k] - 0.7 * e[k-1]

# 定义RLS算法
def rls(phi, y, P, theta, k, lam):
    K = P @ phi / (lam + phi.T @ P @ phi)
    theta = theta + K * (y[k] - phi.T @ theta)
    P = (np.eye(len(P)) - np.outer(K, phi)) @ P / lam
    return theta, P

# 最小方差自校正调节器
def min_variance_self_tuning_controller(y, u, theta, P, k, lam):
    phi_k = np.array([-y[k-1], -y[k-2], u[k-1], u[k-2]])
    theta, P = rls(phi_k, y, P, theta, k, lam)
    u[k] = -(theta[0] * y[k-1] + theta[1] * y[k-2] - theta[2] * u[k-1] - theta[3] * u[k-2])
    return theta, P

# 仿真参数
n = 100  # 仿真长度
m = 2  # 系统时延

# 设置不同的遗忘因子
lam_values = [1, 0.999]

for lam in lam_values:
    # 参数初始化
    theta = np.zeros(4)
    P = np.eye(4) * 1000

    # 重新生成仿真数据
    y = np.zeros(n)
    u = np.zeros(n)
    e = np.random.normal(0, 1, n)

    # 记录参数变化
    theta_history = np.zeros((n, 4))

    # 计算控制输入和系统输出
    for k in range(m, n):
        theta, P = min_variance_self_tuning_controller(y, u, theta, P, k, lam)
        y[k] = system_model(y, u, e, k)
        theta_history[k, :] = theta

    # 绘制系统输出
    plt.plot(y, label='Self-tuned Controlled Output y(k)')
    plt.legend()
    plt.xlabel('k')
    plt.ylabel('y(k)')
    plt.title(f'Controlled System Output with Min Variance Self-tuning Controller, λ={lam}')
    plt.show()

    # 绘制参数变化曲线
    plt.plot(theta_history[:, 0], label='Parameter α1')
    plt.plot(theta_history[:, 1], label='Parameter α2')
    plt.plot(theta_history[:, 2], label='Parameter β0')
    plt.plot(theta_history[:, 3], label='Parameter β1')
    plt.legend()
    plt.xlabel('k')
    plt.ylabel('Parameter Value')
    plt.title(f'Parameter Variation with Self-tuning Controller, λ={lam}')
    plt.show()
```
